```python
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
data_raw = pd.read_csv("data.csv")
data_raw
```

```python
data_final = data_raw.fillna(data_raw.mean(), inplace=False)
data_final
```

```python
data_final.duplicated()
```

```python
mode_data_final = data_final.iloc[:, 1:].mode().iloc[0]
median_data_final = data_final.iloc[:, 1:].median()
print("Mode:")
print(mode_data_final)
print("\nMedian:")
print(median_data_final)
```

```python
histogram_plot1 = sns.histplot(data=data_final, x='fixed acidity')

histogram_plot1.set_title("Histogram plot for fixed acidity")
```

```python
mean_1 = data_final['fixed acidity'].mean()


above_mean_count = data_final[data_final['fixed acidity'] > mean_1]['fixed acidity'].count()
below_mean_count = data_final[data_final['fixed acidity'] < mean_1]['fixed acidity'].count()


labels = ['Greater than Mean', 'Less than Mean']
sizes = [above_mean_count, below_mean_count]
colors = ['yellow', 'violet']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Fixed Acidity Greater than and Less than Mean')
plt.show()
```
